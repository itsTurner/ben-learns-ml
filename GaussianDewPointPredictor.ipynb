##Setting everything up

```
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import math

```

*Open* csv file with Pandas and format into arrays

```
fullWeatherData = pd.read_csv('austin_weather.csv')
weatherData = fullWeatherData[['TempAvgF', 'DewPointAvgF', 'HumidityAvgPercent']]
shuffledWeatherData = weatherData.sample(frac=1)
X_train, X_test, y_train, y_test = train_test_split(shuffledWeatherData[['TempAvgF', 'HumidityAvgPercent']], shuffledWeatherData[['DewPointAvgF']], test_size=0.1, random_state=42)
X_trainArrayTemp = np.array(X_train['TempAvgF'])
X_trainArrayHumidity = np.array(X_train['HumidityAvgPercent'])
X_testArrayTemp = np.array(X_test['TempAvgF'])
X_testArrayHumidity = np.array(X_test['HumidityAvgPercent'])
y_trainArray = np.array(y_train['DewPointAvgF'])
y_testArray = np.array(y_test['DewPointAvgF'])
print(X_trainArrayTemp)
print(X_trainArrayHumidity)
print(X_testArrayTemp)
print(X_testArrayHumidity)
print(y_trainArray)
print(y_testArray)
```

Probability Function

```

exampleArray = []
def probabilityArrayFunc(data, mean, variance):
  probArray = []
  for i in range(len(data)):
    probArray.append((1/(math.sqrt(2*math.pi*variance[i])))*math.exp(-1*((mean[i]-data[i])**2)/(2*variance[i])))
  return probArray


X_testPrediction = []

for j in range(len(X_testArrayHumidity)):
  tempArray = []
  for i in range(len(meanArray)):
    try:
      tempArray.append(np.log(probabilityArrayFunc([X_testArrayTemp[j], X_testArrayHumidity[j]], [meanArray[i][0], meanArray[i][1]], [varianceArray[i][0], varianceArray[i][1]])))
      #break
    except ValueError:
      print("Well this one had 0")      
  X_testPrediction.append(tempArray)
for i in X_testPrediction:
  print(i)
```
