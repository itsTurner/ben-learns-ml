##Setting everything up

```
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import math

```

Open csv file with Pandas and format into arrays

```
fullWeatherData = pd.read_csv('austin_weather.csv')
weatherData = fullWeatherData[['TempAvgF', 'DewPointAvgF', 'HumidityAvgPercent']]
shuffledWeatherData = weatherData.sample(frac=1)
X_train, X_test, y_train, y_test = train_test_split(shuffledWeatherData[['TempAvgF', 'DewPointAvgF']], shuffledWeatherData[['HumidityAvgPercent']], test_size=0.1, random_state=42)
X_trainArrayTemp = np.array(X_train['TempAvgF'])
X_trainArrayDew = np.array(X_train['DewPointAvgF'])
X_testArrayTemp = np.array(X_test['TempAvgF'])
X_testArrayDew = np.array(X_test['DewPointAvgF'])
y_trainArray = np.array(y_train['HumidityAvgPercent'])
y_testArray = np.array(y_test['HumidityAvgPercent'])
print(X_trainArrayTemp)
print(X_trainArrayDew)
print(X_testArrayTemp)
print(X_testArrayDew)
print(y_trainArray)
print(y_testArray)
```
